# 🧠 Single Agent Pipeline Project

## Problem Statement
Build a **Single-Agent Smart Assistant** that:
- Understands user queries
- Routes tasks based on intent
- Uses tools when required
- Returns structured JSON output

### The agent should handle:
- Math queries → Calculator Tool
- Keyword extraction → Keyword Tool
- General queries → Direct response

---
### 🛠️ What You Need to Implement
- Agent logic
- Conditional routing
- Tool integration
- Basic error handling

### 🚀 Bonus

````python
# 🛠️ TOOL 1: Calculator

def calculator(expression: str) -> str:
    """Evaluate a mathematical expression."""
    try:
        return str(eval(expression))
    except Exception:
        return "Error in calculation"
````

````python
# 🛠️ TOOL 2: Keyword Extractor

def extract_keywords(text: str) -> list:
    """Extract keywords from text."""
    try:
        words = text.split()
        keywords = list(set([w.lower() for w in words if len(w) > 4]))
        return keywords[:5]
    except Exception:
        return []
````

## 🤖 Implement Agent Logic Below

👉 Use conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Else → general response

````python
# 🤖 AGENT FUNCTION (IMPLEMENTED)

import re

def agent(query: str):
    # --- Basic input validation ---
    if not isinstance(query, str) or not query.strip():
        return {
            "type": "error",
            "result": "Empty or invalid input received."
        }

    query_lower = query.lower()

    try:
        # --- Route 1: Math queries -> Calculator Tool ---
        if "calculate" in query_lower:
            # Pull out everything after the word "calculate"
            match = re.search(r"calculate\s+(.*)", query_lower)
            expression = match.group(1).strip() if match else ""

            if not expression:
                return {
                    "type": "error",
                    "result": "No mathematical expression found after 'calculate'."
                }

            calc_result = calculator(expression)

            # calculator() returns the string "Error in calculation" on failure
            if calc_result == "Error in calculation":
                return {
                    "type": "error",
                    "result": f"Could not evaluate expression: '{expression}'"
                }

            return {
                "type": "calculation",
                "result": calc_result
            }

        # --- Route 2: Keyword extraction -> Keyword Tool ---
        elif "keywords" in query_lower:
            # Remove the trigger phrase so only the actual text is sent to the tool
            text_for_extraction = re.sub(
                r"extract keywords from|keywords", "", query, flags=re.IGNORECASE
            ).strip()

            if not text_for_extraction:
                return {
                    "type": "error",
                    "result": "No text found to extract keywords from."
                }

            keywords = extract_keywords(text_for_extraction)
            return {
                "type": "keywords",
                "result": keywords
            }

        # --- Route 3: Everything else -> General fallback response ---
        else:
            return {
                "type": "general",
                "result": f"I received your message: '{query.strip()}'. No specific tool matched, so here is a general response."
            }

    except Exception as e:
        # Catch-all so the agent never crashes on an unexpected input
        return {
            "type": "error",
            "result": f"Unexpected error while processing query: {str(e)}"
        }
````

## 📦 Expected Output Format

```
{
  "type": "calculation / keywords / general / error",
  "result": ...
}
```

````python
# 🧪 Test Cases

queries = [
    "Calculate 20 + 5",
    "Extract keywords from Artificial Intelligence is transforming industries",
    "What is machine learning?",
    "Calculate 10 / 0",          # invalid/error case
    "Calculate abc + 2",          # invalid/error case
    "keywords",                    # no text after trigger
    ""                              # empty input
]

for q in queries:
    print("Query:", q)
    print("Response:", agent(q))
    print("-" * 50)
````
